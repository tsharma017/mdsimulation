```python
def dat_to_csv(file_path, flip=False):
    # List of lipids to flip
    lipids_to_flip = ["POPI", "POPS", "POPE", "POPC", "PSM", "CHL", "POPI24"]

    # Open the .dat file for reading
    with open(file_path, 'r') as file:
        # Read the contents of the file into a variable
        file_contents = file.read()
    
    # Split the contents of the file by newline character
    lines = file_contents.split("\n")
    
    # Write the contents to a .csv file with the same name as the .dat file
    with open(file_path.replace(".dat", ".csv"), 'w') as file:
        for line in lines:
            # Skip the line if it is empty
            if line == "":
                continue
                
            # Split the line into columns
            columns = line.split()
            
            # Swap the donor and acceptor columns if flip is True and the first column starts with the desired lipids
            if flip and any(map(lambda x: columns[0].startswith(x), lipids_to_flip)):
                columns[0], columns[1] = columns[1], columns[0]
            
            # Write each line as a row in the .csv file
            file.write(",".join(columns) + "\n")

# Example usage with flip
dat_to_csv('1420620hbonds-details.dat', True)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

# Define a linear function for fitting
def linear_function(x, a, b):
    return a * x + b

# Load MSD data
df = pd.read_csv('DDdiff2.csv')
time_ns = df['ns']
msd = df['DD']

# Calculate MSD for incremental tau values
max_tau = int(time_ns.iloc[-1])  # Maximum value of tau
tau_values = np.arange(1, max_tau + 1)  # Tau values from 1 to max_tau
msd_values = []

for tau in tau_values:
    displacements = []
    for i in range(len(time_ns) - tau):
        displacements.append(msd[i + tau] - msd[i])
    msd_tau = np.mean(np.square(displacements))
    msd_values.append(msd_tau)

# Fit the linear function to the MSD vs. tau data
popt, _ = curve_fit(linear_function, tau_values, msd_values)
a, b = popt
D = a / 6  # Divided by 6 for 3D diffusion (2nD, n=3)

# Format D in exponential notation
D_formatted = f'{D:.2e} m²/s'  # Assuming D is in m²/s

# Range for the slope line (50 ns to 400 ns)
slope_tau_range = np.arange(50, 401)  # 401 because the upper bound is exclusive in np.arange
slope_msd_values = linear_function(slope_tau_range, a, b)

# Plotting
plt.plot(tau_values, msd_values)
plt.plot(slope_tau_range, slope_msd_values, linestyle='--', color='red')

plt.xlabel('$\\tau$ (ns)', fontsize = 22)
plt.ylabel('MSD (Å²)', fontsize = 22)
plt.yticks(range(0, 101, 25))
plt.tick_params(axis='both', which='major', labelsize=19)
#plt.legend()

# Annotate the slope value at the midpoint of the specified range
mid_slope_tau = (slope_tau_range[0] + slope_tau_range[-1]) // 2
mid_slope_msd = linear_function(mid_slope_tau, a, b)
#plt.annotate(f'Slope: {a:.2f}', xy=(mid_slope_tau, mid_slope_msd), xytext=(mid_slope_tau, mid_slope_msd*1.2))

# Show the plot
plt.show()

print(f'Diffusion Coefficient (D): {D_formatted}')
```
